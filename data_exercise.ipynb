```python
nrow, ncol = df_sorted.shape
```

```python
f1c4 = open('1c4.txt', 'w')
f2s0 = open('2s0.txt', 'w')
f4c1 = open('4c1.txt', 'w')

flist = [f1c4, f2s0, f4c1]
```

```python
# extracting the numbers to the respective files based on the reference range provided

for ik in range(nrow):
    v = df_sorted.iloc[ik, :]
    
    if(0<  v[0] < 360) and (150 < v[1]<180):
        flist[0].write('{} {} {}\n'.format(v[0], v[1], v[2]))
    
    elif(135<  v[0] < 165) and (75 < v[1] < 105):
        flist[1].write('{} {} {}\n'.format(v[0], v[1], v[2]))
    
    elif(0<  v[0] < 360) and (0 < v[1]<30):
        flist[1].write('{} {} {}\n'.format(v[0], v[1], v[2]))
    else:
        print("not belong to any group")
```

```python
for f in flist:
    f.close()
```
